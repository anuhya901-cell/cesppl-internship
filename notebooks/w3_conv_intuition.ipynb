Use scipy.signal.convolve2d
to manually apply three different kernels to a grayscale image of your choice: an
edge detector (the Sobel kernel from Ng's lecture), a blur (a 3x3 array of ones
divided by 9), and an identity (all zeros except a 1 in the centre). Display the four
images side by side: original and three filtered.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
```

Read Image as Grayscale

```python
img = cv2.imread("img2.jpg", cv2.IMREAD_GRAYSCALE)

print(img.shape)

plt.imshow(img, cmap="gray")
plt.show()
```

Create Three Kernels

1. Sobel Edge Detector

```python
sobel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])
```

2. Blur Kernel

```python
blur = np.ones((3,3)) / 9
```

3. Identity Kernel

```python
identity = np.array([
    [0,0,0],
    [0,1,0],
    [0,0,0]
])
```

Apply Convolution

```python
edge_img = convolve2d(img, sobel, mode="same")

blur_img = convolve2d(img, blur, mode="same")

identity_img = convolve2d(img, identity, mode="same")
```

Display Results

```python
plt.figure(figsize=(15,5))

plt.subplot(1,4,1)
plt.imshow(img, cmap="gray")
plt.title("Original")

plt.subplot(1,4,2)
plt.imshow(edge_img, cmap="gray")
plt.title("Edge Detection")

plt.subplot(1,4,3)
plt.imshow(blur_img, cmap="gray")
plt.title("Blur")

plt.subplot(1,4,4)
plt.imshow(identity_img, cmap="gray")
plt.title("Identity")

plt.show()
```

Edge Detection Kernel

The Sobel kernel highlights regions where pixel values change rapidly. These changes usually correspond to edges of objects. CNNs use similar filters in early layers to detect edges and shapes.

Blur Kernel

The blur kernel averages neighboring pixels. This smooths the image and reduces noise. It removes fine details while preserving the overall structure.

Identity Kernel

The identity kernel keeps the center pixel unchanged and ignores surrounding pixels. The output image remains almost identical to the input image.

Feature Extraction

Different kernels extract different information from the same image. CNNs learn these filters automatically during training.

Why CNNs Use Filters

Filters allow CNNs to detect useful visual patterns such as edges, corners, textures, and object parts. These learned features help the network recognize complex objects.
